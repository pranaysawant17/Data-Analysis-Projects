##Q3

Converting data from Schedule Days column to numrical value to match given output

```python
y = {'monday':0,'tuesday':1, 'wednesday':2,'thursday':3, 'friday':4, 'saturday':5, 'sunday':6}
```

```python
def convertion(list_):
 
  if len(list_)==1:
    for i in list_:
      return str(y[i.lower()])
  else:
    lst=[]
    for i in list_:
      lst.append(y[i.lower()])
      
    return  str(lst)
```

```python
spb['Schedule Days']=spb['Schedule Days'].apply(eval)
spb['Schedule Days']=spb['Schedule Days'].apply(convertion)
spb['Schedule Days']=spb['Schedule Days'].str.strip('[]')
spb
```
